Why We use generators: 
1. Memory efficient
2. Stats from where it left before. --> Keeps track of previous *state*
3. Execution --> Execute all at once(funtion) --> Pauses function(lazy evaluation)
4. Gets one result at a time and not multiple values

```python
def get_squares_list(n):
    result = []
    for i in range(n):
        result.append(i ** 2)
    return result

def get_squares_gen(n):
    for i in range(n):
        yield i ** 2

import sys

list_result = get_squares_list(100)
print(list_result)
gen_result = get_squares_gen(100)
print(list(gen_result))

print(f"{sys.getsizeof(list_result)} bytes")
print(f"{sys.getsizeof(gen_result)} bytes")

#Generator uses very less memory
```

| Regular Function (`return`)         | Generator (`yield`)                                     |
| ----------------------------------- | ------------------------------------------------------- |
| Ends the function immediately       | Pauses the function                                     |
| Returns one value                   | Can produce multiple values over time                   |
| Doesn't remember its state          | Remembers where it stopped                              |
| Executes once                       | Resumes each time `next()` is called                    |
| Uses more memory for large datasets | Memory-efficient because values are generated on demand |

```python
def count_up_to(n):
    i = 0
    while i < n:
        yield i
        i += 1

counter = count_up_to(5)
print(next(counter)) #-> 0 (state)
print(next(counter)) #--> 1 (State)
print(next(counter))
print(next(counter))
print(next(counter))


# #Loop 
# for num in count_up_to(5):
#     print(num)
```

```python
def fibonaaci(limit):
    """Generate Fibonacci numbers up to limit"""
    a, b = 0, 1
    while a < limit:
        yeild a 
        a, b = b, a + b

print("Fibonacci sequence upto 1000 limit: ")
for num in fibonaaci(1000):
    print(num) #next line
print()
```

```python
#Infinite generator

def infinite_generator():
    """Generate numbers forever"""
    i = 0
    while True:
        yield i
        i += 1

counter = infinite_generator()
print("First 10 from infinite generator")
for _ in range(10):
    print(next(counter), end=" ")
print()
```

```python
# Generator with conditions

def even_numbers_generator(limit):
    for i in range(limit):
        if i % 2 == 0:
            yield i

print("Even number list")
for num in even_numbers_generator(20):
    print(num, end=" ")
print()
```

```python
# Batch processing:

events = list(range(1, 101))

#without generators
def process_batch_list(data, batch_size=10):
    batches = []
    for i in range(0, len(data), batch_size):
        print(f"Creating batch {i//batch_size}")
        batches.append(data[i:i+batch_size])
    return batches
for batch in process_batch_list(events):
    print(f"Processing bacth {batch}")
```

```python
#With Generator
def batch_generator(data, batch_size=10): #Parameter
    for i in range(0, len(data), batch_size):
      print(f"Creating batch {i//batch_size}")
      yield data[i:i+batch_size]
                

for batch in batch_generator(events, 10): #Arguements
    print(f"Processing batch: {batch}")
```

# Decorators

What are Decorators?

Decorator is a function that 
1. Takes another function as input
2. Adds new funtionality to it
3. Returns a modified version
4. No Need to change original function.


Synytax: 

def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@my_decorator
def say_hello():
    print("Hello")

say_hello()


Output:
Before
Hello
After

person without clothes == Original Function
person wearing clothes == Decorated Function

```python
def hello_decorator(func):
    def wrapper():
        print("Function is called") #-- time.time() -- Start time
        func()
        print("Function Completed") #--> time.time() -- End time 
        # total_time = End_time - start_time
    return wrapper

@hello_decorator
def say_hello():
    print("Hello")

say_hello()
```

```python
# Decorator with Arguements
                    #greet(name)
def greet_decorator(func):
    def wrapper(name): #-- > "Rahul" 
        print(f"🎯 Calling function for {name}")
        result = func(name) #--> greet("Rahul")
        print(f"✅ Function completed for {name}") #--> Rahul
    return wrapper

@greet_decorator
def greet(name):
    print(f"Hello, {name}!")
    # return f"Greeted {name}"

greet("Rahul")
```

```python
#Timer Decorator
import time
def timer_decorator(func):
    def wrapper(seconds):
        start_time = time.time()
        print(f" Starting funtion: {func.__name__}") #Dunder Methods

        result = func(seconds)

        end_time = time.time()
        duration = end_time - start_time

        print(f"Completed function in{duration:.4f} seconds")
        return result
    return wrapper


@timer_decorator
def slow_function1(seconds):
    time.sleep(seconds)
    print(f"Function execution")

slow_function1(seconds=10)
```

```python
#Logger Decorator

def logger_decorator(func):
    def wrapper(*args, **kwargs):
        print(f"Function Log call")
        print(f" Function: {func.__name__}")
        print(f"Arguments: {args}")
        print(f"Keyword Arguments: {kwargs}")

        result = func(*args, **kwargs)

        print(f"Result: {result}")
        print("Log Complete")

        return result
    return wrapper

@timer_decorator
@logger_decorator
def add(a, b):
    return a + b 

@logger_decorator
def subtract(a, b):
    return a - b 

# logger_decorator(add)

add(1500, 3000)
subtract(3000, 1500)
```

```python
#Caching Decorator

def cache_decorator(func):
        cache = {}
        def wrapper(n):
                if n in cache:
                        print(f"eReturning the cached value for {n}")
                        return cache[n]

                print(f"Computing for {n}")
                result = func(n)
                cache[n] = result 
                return result
        wrapper.cache = cache
        return wrapper                       



@cache_decorator
def expensive_computation(n):
        """Simulates expensive operation"""
        time.sleep(0.5)
        return n ** 2

print(expensive_computation.cache)
print(expensive_computation(5))
print(expensive_computation(5))
print(expensive_computation(10))
print(expensive_computation(5))
print(expensive_computation(20))
print(expensive_computation(10))
print(expensive_computation.cache)
```

```python
#Multiple Decorators
def decorator1(func):
    def wrapper(*args, **kwargs):
        print("1️⃣  Decorator 1 - Before")
        result = func(*args, **kwargs)
        print("1️⃣  Decorator 1 - After")
        return result
    return wrapper
def decorator2(func):
    def wrapper(*args, **kwargs):
        print("2️⃣  Decorator 2 - Before")
        result = func(*args, **kwargs)
        print("2️⃣  Decorator 2 - After")
        return result
    return wrapper

@decorator1
@decorator2
def my_function():
    print("🔵 Main function")

my_function()
```

```python
#Permission Decorator -- Very important
def require_admin(func):
    def wrapper(user_role, *args, **kwargs):
        if user_role != "admin":
            raise PermissionError(f"❌ {user_role} not allowed! Admin only!")
        
        print(f"✅ Admin access granted")
        return func(*args, **kwargs)
    
    return wrapper

@require_admin
def delete_database():
    return "Database deleted!"


print(delete_database("admin"))

try:
    print(delete_database("user"))
except PermissionError as e:
    print(f"Error: {e}")
```
