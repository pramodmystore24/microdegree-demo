```python
def calculate_weekly_money(daily_amount):
    weekly = daily_amount * 7
    print(f"Daily: Rs {daily_amount}")
    print(f"Weekly: Rs {weekly}")

calculate_weekly_money(50)
calculate_weekly_money(100)
calculate_weekly_money(500)
```

```python
def calculate_age(birth_year):
    current_year = 2026
    age = current_year - birth_year
    print(f"You are {age} years old ")

```
