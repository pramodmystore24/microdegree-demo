```python
def calculate_discount(price, discount_percent):
    discount_amount = price * discount_percent / 100
    final_price = price - discount_amount
    return final_price
```

```python
original = 1000
discounted = calculate_discount(original, 20)
print(f"Original: {original}")
print(f"Dicounted price: {discounted}")
```

```python
total_items = 5
total_cost = calculate_discount(original, 15) * total_items
print(f"Total buying cost for 5 pairs with 15% of Shoes is {total_cost}")
```

```python
def is_valid_age(age):
    if age > 0 and age < 80:
        return True
    else:
        return False
    
if is_valid_age(200):
    print("Age is valid")
else:
    print("Invalid age")

```

```python
ages = [15, -5, 120, 45, 0, 98]
for age in ages:
    if is_valid_age(age):
        print(f"{age} is valid")
    else:
        print(f"{age} is invalid")
```

```python
def is_valid_email(email):

    if "@" in email and "." in email:
        return True
    else:
        return False
    
emails = ["lakshmikanth@gmail.com", "invalid.email", "test@gmail", "pramod@gmail.com"]
for each_email in emails:
    if is_valid_email(each_email):
        print(f"{each_email} is valid")
    else:
        print(f"{each_email} is invalid")

```

```python
def sum_all(numbers):
    total = 0
    for num in numbers:
        total += num
    return total

scores = [95,85 ,55 ,65, 75, 78]
total = sum_all(scores)
print(f"Total Score {total}")
print(f"Average Score {total/ len(scores)}")
```

```python
def find_highest(scores):
    highest = 0
    for score in scores:
        if score > highest:
            highest = score
    return highest

math_scores = [95,85 ,55 ,65, 75, 78]
english_scores = [95,88 ,56 ,65, 98, 78]

print(f"Highest Math Score is {find_highest(math_scores)}")
print(f"Highest English Score is {find_highest(english_scores)}")
```

```python
def calculate_average(marks):
    return sum(marks) /len(marks)

def get_grade(average):
    if average >= 90:
        return "A"
    elif average >= 80:
        return "B"
    elif average >= 70:
        return "C"
    elif average >= 60:
        return "D"
    else:
        return "F"
    
def is_passing(marks):

    for mark in marks:
        if mark < 60:
            return False
    return True

def compare_students(student_data):
    print()
    print("=" * 40)
    print("Class Performance Report")
    print("=" * 40)

    topper = ""
    highest_score = 0

    for name, marks in student_data.items():

        average = calculate_average(marks)
        grade = get_grade(average)

        if is_passing(marks):
            status = "Pass"
        else:
            status = "Fail"
        
        print(f"{name} Average: {average} Grade: {grade} Status: {status}")

        if average > highest_score:
            highest_score = average
            topper = name
    print("=" * 40)
    print(f"Topper: {topper}")
    print(f"Highest Score: {highest_score}")
    print("=" * 40)

all_students = {
    "Rahul": [92, 85, 88, 90, 87],
    "Priya": [78, 82, 80, 79, 81],
    "Arjun": [95, 93, 97, 94, 96],
    "Sneha": [68, 70, 72, 69, 71],
    "Karan": [95, 45, 88, 91, 84]
}

compare_students(all_students)

```

```python
#Arguements --> Args --> *args
#Keyword Arguments --> Kwargs --> **kwargs

def print_items(*items):
    """Prints items with their index12345""" #DocString
    for i, item in enumerate(items, start=1):
        print(f"{i}. {item}")
```

```python
def create_profile(**details):
    """Create a profile with any information"""
    print("User Profile")
    print("=" *40)

    for key, value in details.items():
        display_key = key.replace("_", " ").title()
        print(f"{display_key}: {value}")
    print("=" *40)
```

```python
create_profile(name = "Rahul", age = 25, city= "Bengaluru")
create_profile(name = "Pramod", age = 25, city= "Delhi", job="AI Engineer", hobby= "Writing Code")
create_profile(name = "Pramod", age = 25, city= "Delhi", job="AI Engineer", hobby= "Writing Code", depatment= "IT")
```

Return Values --> return value --> Function whose result can be used further
*args --> def f_name(*args) --> Multiple input arguments
