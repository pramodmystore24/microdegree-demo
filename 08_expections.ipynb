```python
def divide_numbers(a, b):
    try:
        result = a/b
        print(f"{a} / {b} = {result}")
    except ZeroDivisionError:
        print(f"Cannot divide by Zero")
    except TypeError:
        print(f"Both inputs must be numbers")
```

```python
divide_numbers(10, 2)
divide_numbers(10, 0)
divide_numbers(10, "5")
```

```python
def pocess_student_data(data):
    try:
        name = data["name"]
        age = int(data["age"])
        return f"{name} is {age} years old"
    except KeyError as e:
        print(f"Error Details: {e}")
    except ValueError as e:
        print(f"Value Error Details: {e}")
    except Exception as e:
        print(f"unexpected error Details: {type(e).__name__}: {e}")
```

```python
def validate_age(age):
    try: 
        age = int(age)
    except ValueError:
        raise ValueError(f"Age must be number, not {age}")

    if age < 0 or age > 100:
        raise ValueError(f"{age} is not valid, must be 0 to 100")
    return age

test_ages = ["25", "-5", "200", "abc"]

for age in test_ages:
    try:
        valid_age = validate_age(age)
        print(f"Age {valid_age} is valid")
    except ValueError as e:
        print(f"Error Details: {e}")
```

```python
def withdraw_money(balance, amount):
    if not isinstance(balance, (int, float)):
        raise TypeError("Balance must be a number")
    if not isinstance(amount, (int, float)):
        raise TypeError("Amount must be a number")
    if amount <= 0:
        raise ValueError("Withdrwal amount should be positive")
    if amount >  balance:
        raise ValueError("Insufficient balance")
    
    return balance - amount


test_cases = [
    (5000, 2000),
    (5000, 8000),
    (5000, -100),
    ("5000", 1000)
]

for balance, amount in test_cases:
    try:
        new_balance = withdraw_money(balance, amount)
        print(f"Withdraw of Rs {amount} is successful and balance is Rs {new_balance}")
    except (ValueError, TypeError) as e:
        print(f"{e}")
```

```python
def read_file_safe(filename):
    try:
        with open(filename, "r") as f:
            return f.read()
    except FileNotFoundError:
        print("File do not exists")
    except PermissionError:
        print("No Permission to read ")
    

print("Reading the Existing file:")
content = read_file_safe("/Users/pramodramamurthy/Downloads/Micrdegree_AI/notes_guru1.txt")
if content:
    print(f"{content[:50]} ....")
else:
    print(f"File not exist --> else block")
```
