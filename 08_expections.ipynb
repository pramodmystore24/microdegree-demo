```python
def get_ur_age():
    try:
        age = int(input("Enter your age: "))
        if age < 0 or age > 100:
            print(f"Age {age} must be between 0 to 100")
        else:
            print(f"{age} is valid")
    except ValueError:
        print(f"Enter a valid number")

get_ur_age()

```

```python
def divide_numbers(a, b):
    try:
        result = a/b
        print(f"{a} / {b} = {result}")
    except ZeroDivisionError:
        print(f"Cannot divide by Zero")
    except TypeError:
        print(f"Both inputs must be numbers")
```

```python
divide_numbers(10, 2)
divide_numbers(10, 0)
divide_numbers(10, "5")
```

```python
def pocess_student_data(data):
    try:
        name = data["name"]
        age = int(data["age"])
        return f"{name} is {age} years old"
    except KeyError as e:
        print(f"Error Details: {e}")
    except ValueError as e:
        print(f"Value Error Details: {e}")
    except Exception as e:
        print(f"unexpected error Details: {type(e).__name__}: {e}")
```

```python
pocess_student_data({"name": "Rahul", "age": "25"})
```

```python
pocess_student_data({"name": "Rahul"})
```

```python
def validate_age(age):
    try: 
        age = int(age)
    except ValueError:
        raise ValueError(f"Age must be number, not {age}")

    if age < 0 or age > 100:
        raise ValueError(f"{age} is not valid, must be 0 to 100")
    return age

test_ages = ["25", "-5", "200", "abc"]

for age in test_ages:
    try:
        valid_age = validate_age(age)
        print(f"Age {valid_age} is valid")
    except ValueError as e:
        print(f"Error Details: {e}")
```
