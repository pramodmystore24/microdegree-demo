```python
number = 42
text = "1000"
decimal = 3.994567

print(str(number)) # 42 --> "42" (int to string)
print(int(text)) # "1000" --> 1000(string to int)
print(float(number)) #42 --> 42.0 (int to float)
print(int(decimal)) #3.994567 --> 3(float to int(whole number))

print(bool(0)) # 0 --> False 
print(bool(1)) #1 --> True (any non-zero is True)

print(bool("hello")) # "hello" --> True (non empty string is always True)
print(bool("")) #Empty String "" --> False (empty string is always False)
```

```python
#Arithmetic Operations

total_runs = 12045
total_innings = 120
not_outs = 50
total_match = 250 

#Addition 
new_score = total_runs + 500
print("Total Runs : ",new_score, "Total innings" ,total_innings)
print(f"Total Runs : {new_score}, Total innings: {total_innings} ,Not Outs: {not_outs}, Total Match: {total_match}")


#Subtraction
completed_innings = total_innings - not_outs
print("Completed Innings: ", completed_innings)

#Multiplication
print("Prize Money: ",total_match * 10000)

#Divison -> always give you float/decimal values --> decimal 

batting_avg = total_runs /(total_innings - not_outs)
print(f"Batting average:  {batting_avg:.2f}")

#//Floor Division --> Ingteger --
ordinary_division = total_runs / 100
full_centuries = total_runs // 100 
print(f"ordinary Division: {ordinary_division}")
print(f"Roughly how many full centuries player has player: {full_centuries}")

#Exponent --> power --> 2 to the power of 100 
#  2 ** 10 

print(f"Two to the power of ten is {2 ** 10}")
# 2*2*2*2.... 10 times 



#Formatted Print/ Output 
```
