```python
#Indexing & String Length

#Index 
name = "Dhoni"
#Index  01234. (If you read it from left index starts from 0) -- end is not defined
#Index. -9-8-7-6-5-4-3-2-10 (If you read it from right index starts from -1)

length_name = len(name)

print(f"Full name {name}")
print(f"Length = {length_name} chars")

#Access individual characters
print(f"First Char is {name[0]}") 
print(f"Last Char is {name[-1]}")
print(f"Middle {name[2]}")



```

```python
# Slicing of the string -- Bread slice 

movie_title = "Sholay (1975) - Bollwood Classic"
#Indexing      0123456789......

#[start:end] --> (end is not included)
title_only = movie_title[0:6]
year_only = movie_title[8:12]
description = movie_title[16:]
print(title_only)
print(year_only)
print(description)

```

```python
phone_number = "+91 9876543210"
country_code = phone_number[1:3]
number_part = phone_number[4:]
last_4 = phone_number[-4:]
print(country_code)
print(number_part)
print(last_4)
```

```python
#slicing -- [start:end:step]
word = "BOLLYWOOD"
every_other_word = word[::3]
print(every_other_word)

reversed_word = word[::-1]
print(reversed_word)
```

```python
sentence = "Sachin Tendulkar is the God of Cricket"

print(sentence.find("God"))       # 30  (position where "God" starts)
print(sentence.find("Kohli"))     # -1  (not found → returns -1)
print(sentence.count("i"))        # 3   (how many times "i" appears)
print(sentence.startswith("Sachin"))  # True
print(sentence.endswith("Hockey"))    # False
print("Cricket" in sentence)          # True  (check if word exists)
```
