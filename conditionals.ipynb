```python
print("=" * 40)
print(" 🏦 Welcome to the Python Bank")
print("=" * 40)

accounts = {
    "123456" : {"pin":"12345", "balance": 15000, "name": "Rahul"},
    "789012" : {"pin":"5678", "balance": 8000, "name": "Priya"}
}

card_number = input("\n Enter Card Number: ").strip()

if card_number in accounts:
    account = accounts[card_number]

    #Pin check (3 attempts)
    attempts = 0
    pin_correct = False

    while attempts < 3:
        pin = input("Enter PIN: ").strip()
        if pin == account["pin"]:
            pin_correct = True
            break
        else:
            attempts += 1
            remaining = 3 - attempts
            if remaining > 0 :  #Remaining Attempts
                print(f"Wrong Pin {remaining} attempts remaining")
            else:
                print("Card Blocked. Too many attempts")
    
    if pin_correct: 
        print(f"Welcome {account['name']}")

        while True:
            print("\n 1. Check Balance")
            print("2. Withdraw Cash")
            print("3. Depocit Cash")
            print("4. Exit")

            choice = input("Select option: ").strip()

            if choice == "1":
                print(f"Available balance{account['balance']}: ")

            elif choice == "2":
                amount_str = input("Enter the amount you want to withdraw(multiple of 100 only): ")
                
                if not amount_str.isdigit():
                    print("Invalid entry. Numbers only")
                else:
                    amount = int(amount_str)

                    if amount <= 0:
                        print("Amount must be some positive")
                    elif amount %100 !=0:
                        print("Must be in multiples of 100")
                    elif amount > 10000:
                        print("Max 10,000 allowed per transaction")
                    elif amount > account["balance"]:
                        print("Inssufficent amount")
                    else: 
                        account["balance"] -= amount
                        print(f"Amount {amount} can be dispensed Remaining: {account["balance"]}")

            elif choice == "3":
                deposit = input("Enter the amount you want to deposit(multiple of 100 only): ")
                if deposit.isdigit() and int(deposit) > 0:
                    account["balance"] += int(deposit)
                    print(f"Rs{int(deposit)} is deposited. New balance is {account["balance"]}")

            elif choice == "4":
                print("Thanks for using Python ATM ")
                break
            else:
                print("Invalid Option.")

else:
    print("Card unrecognized")






            
            

```

```python
# #Match Case


# if day_num == 1:
#     print("day is Monday")
# elif day_num == 2:
#     print("day is Tuesday")
# elif day_num == 3:
#     print("day is Wednesday")
# elif day_num == 3:
#     print("day is Thursday")



# else: 
# .. etc

# Cleaner multiple choice conditionals
day_num = int(input("Enter the day of the week(1-7)"))
match day_num:
    case 1:
        day = "Monday"
    case 2:
        day = "Tuesday"
    case 3:
        day = "Wed"
    case 4:
        day = "Thur"
    case 5:
        day = "Fri"
    case 6:
        day = "Sat"
    case 7:
        day = "Sun"
    case _:
        day = "Invalida day"

print(f"Day: {day}")
```
