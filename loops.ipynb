```python
subjects = ["Math", "Science", "English", "Computer"]
marks = [95, 82, 76, 89]

print("=" * 40)
print("Report Card")
print("=" * 40)

total = 0


for subject, mark in zip(subjects, marks):
    print(f"{subject} {mark}")
    total += mark

average = total / len(marks)

print("-"* 40)
print(f"Average: {average}")

if average >= 90:
    grade = "A+"
elif average >= 70:
    grade = "A"
elif average >= 60:
    grade = "B"
elif average >= 40:
    grade = "C"
else:
    grade = "D"

print(f"Grade:{grade}")
```

```python
#Break Statement Continue Pass
scores = [85, 72, 91, 38, 88, 43, 95]

for num, each_score in enumerate(scores):
    if each_score < 40:
        print(f" Student {num+1} failed with {each_score}.Stop searching ")
        break
    print(f"Student {num+1}: {each_score} --> Pass")
```

```python
#Continue
scores = [85, 72, 91, 38, 88, 43, 95]
for num, each_score in enumerate(scores):
    if each_score < 40:
        continue #Skip and continue forward
    print(f"Student {num+1}: {each_score} --> Pass")
```

```python
#Nested Loops -- Loops inside a loop 
import random

teams = {
    "Mumbai Indians": ["Rohit", "Gill", "Virat", "Iyer", "Hardik"],
    "Chennai Super Kings": ["Ruturaj", "Conway", "Dhoni", "Jadeja", "Dube"]
}

print("IPL Match - Innings Score Card")
#Team Name
#Batting Order
#Over 1: 6 W 0 W 1 4 --> Total Score
# Over 2: 6 W 0 W 1 4 --> Total Score
# Over 3: 6 W 0 W 1 4 --> Total Score
# Over 4: 6 W 0 W 1 4 --> Total Score
# Over 5: 6 W 0 W 1 4 --> Total Score
 #total Score
 #average Score

for team_name, players in teams.items():

    total_runs = 0
    wickets = 0

    print(f"{team_name.upper()} Battting")
    print("-"* 50)

    print("Batting Order")
    for player in players:
        print(player)

    print("-"* 50)

    for over in range(1, 6):

        over_runs = 0
        balls = []

        for ball in range(1, 7):

            outcome = random.choice([0, 0, 1, 2, 4, 6, "W", "W" ])
            balls.append(str(outcome))

            if outcome == "W":
                wickets += 1
            else:
                over_runs += outcome

        total_runs += over_runs

        print(f" Over {over}: {' '.join(balls)} --> {over_runs} runs")
    
    print("-"* 50)
    print(f"Total Score: {total_runs}/{wickets}")
    print(f"Average: {total_runs/5}")






```
