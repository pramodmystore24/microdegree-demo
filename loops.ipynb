```python
items = ["Rice", "Milk", "Bread", "EggsEggseg", "Oil"]
prices = [60, 35, 40, 72]

print("=" * 40)
print(" Shopping Bill")
print("=" * 40)

total = 0

for each_item, each_price in zip(items, prices):
    print(f"{each_item:<10}--> Rs{each_price:<10}")
    total += each_price

print("-" * 40)
print(f"Total Bill: Rs.{sum(prices)}")

```

```python
subjects = ["Math", "Science", "English", "Computer"]
marks = [95, 82, 76, 89]

print("=" * 40)
print("Report Card")
print("=" * 40)

total = 0


for subject, mark in zip(subjects, marks):
    print(f"{subject} {mark}")
    total += mark

average = total / len(marks)

print("-"* 40)
print(f"Average: {average}")

if average >= 90:
    grade = "A+"
elif average >= 70:
    grade = "A"
elif average >= 60:
    grade = "B"
elif average >= 40:
    grade = "C"
else:
    grade = "D"

print(f"Grade:{grade}")
```

```python
employees = {
    "Rahul": 45000,
    "Anita": 52000,
    "Vijay": 61000,
    "Priya": 49000
}

print("=" * 40)
print("Salary Card")
print("=" * 40)

for name, salary in  employees.items():
    
    bonus = salary * 0.10
    total = salary + bonus

    print() #Alternative for new line
    print(f"\nEmployee: {name}")
    print(f"Salary: {salary}")
    print(f"Bonus: {bonus}")
    print(f"Total: {total}")

```

```python
players = {
    "Virat": {"Runs": 800, "Average": 54},
    "Gill": {"Runs": 680, "Average": 48},
    "Rohit": {"Runs": 590, "Average": 43},
    "KL Rahul": {"Runs": 800, "Average": 65}
}

print("=" * 40)
print("Cricket Stats")
print("=" * 40)

highest = 0
best_player = ""

for player, stats in players.items():
    print(player)
    print(f"Highest Runs: {stats["Runs"]}")
    print(f"Average: {stats["Average"]}")
    print()
        
    if stats["Runs"] > highest:
        highest = stats["Runs"]
        best_player = player

print("=" * 40)
print(f"Highest Runs Scorer: {best_player}")
print(f"Runs: {highest}")
```
