```python
#Nested Dictionary

# Dictionary --> Dictonary
students_db = {
    "STU001": {
        "name"  : "Rahul Gupta",
        "marks" : {"Maths": 88, "Science": 92, "English": 79},
        "grade" : "A",
    },
    "STU002": {
        "name"  : "Sneha Patel",
        "marks" : {"Maths": 95, "Science": 97, "English": 91},
        "grade" : "A+",
    },
    "STU003": {
        "name"  : "Arjun Singh",
        "marks" : {"Maths": 62, "Science": 70, "English": 68},
        "grade" : "B",
    }
}


# print(f"{students_db["STU001"]["name"]}: {students_db["STU001"]["marks"]["Maths"]}")
# print(students_db["STU002"]["marks"]["Maths"])
# print(students_db["STU002"]["marks"].get("History","N/A"))


print("Student Report Card")

for student_id, info in students_db.items():
    name = info["name"]
    marks = info["marks"]
    grade = info["grade"]

    total = sum(marks.values())
    average = total/ len(marks)
    
    print(f" \n{student_id} - {name} [{grade}]")
    for subject, score in marks.items():
        bar = "❚" * (score // 10)
        print(f"{subject}: {score}  {bar}")
    print(f"\n Average :{average:.1f}")




```

```python
team_india = {"Rohit", "Virat", "Dhoni", "Bumrah", "Jadeja"}
team_rcb = {"Virat", "ABD", "Chris Gayle", "Rajat"}
team_mi = {"Rohit", "Bumrah", "Malinga", "Pollard"} 

#Intersection - palyers in Both the teams
in_both = team_india & team_rcb 
print(f"In Both :{in_both}")

#Difference  -- In India but not in RCB
only_india = team_india - team_rcb
print(f"Only in India :{only_india}")

#Union -- Players in either team (combined)
all_players = team_india | team_rcb 
print(f"All Players :{all_players}")

#Symmetric Difference -- in one but not in both
exclusive = team_india ^ team_rcb
print(f"Exclusive : {exclusive}")
```

```python
users = {
    "admin": "admin123",
    "pramod": "python123"
}

while True:

    print("\n🔐 Login System")
    print("-" * 30)
    print("1. Register User")
    print("2. Login")
    print("3. Change Password")
    print("4. Delete User")
    print("5. Show All Users")
    print("6. Exit")

    choice = input("\nEnter your choice: ")

    # ---------------- Register ----------------
    if choice == "1":

        username = input("Enter username: ")

        if username in users:
            print("❌ Username already exists.")
        else:
            password = input("Enter password: ")
            users[username] = password
            print("✅ User registered successfully.")

    # ---------------- Login ----------------
    elif choice == "2":

        username = input("Enter username: ")
        password = input("Enter password: ")

        if username in users:

            if users[username] == password:
                print(f"✅ Welcome {username}!")
            else:
                print("❌ Incorrect password.")

        else:
            print("❌ User not found.")

    # ---------------- Change Password ----------------
    elif choice == "3":

        username = input("Enter username: ")

        if username in users:

            old_password = input("Enter current password: ")

            if users[username] == old_password:

                new_password = input("Enter new password: ")
                users[username] = new_password

                print("✅ Password updated successfully.")

            else:
                print("❌ Incorrect current password.")

        else:
            print("❌ User not found.")

    # ---------------- Delete User ----------------
    elif choice == "4":

        username = input("Enter username: ")

        if username in users:

            confirm = input("Are you sure? (yes/no): ").lower()

            if confirm == "yes":
                del users[username]
                print("✅ User deleted successfully.")
            else:
                print("Deletion cancelled.")

        else:
            print("❌ User not found.")

    # ---------------- Show Users ----------------
    elif choice == "5":

        if users:

            print("\nRegistered Users")
            print("-" * 30)

            for username in users:
                print(username)

        else:
            print("No users registered.")

    # ---------------- Exit ----------------
    elif choice == "6":

        print("👋 Thank you for using the Login System!")
        break

    else:
        print("❌ Invalid choice.")
```
