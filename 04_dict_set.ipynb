```python
student = {
    "name"    : "Priya Sharma",
    "age"     : 21,
    "college" : "Delhi University",
    "cgpa"    : 8.7,
    "passed"  : True
}

#Method1:

# print(student["name"])

# #Method2
# # get()
print(student.get("name"))
print(student.get("phone"))
print(student.get("phone", "Phone number not present"))

#Modifycation
student["cgpa"] = 9.1
print(f"Updated CGPA: {student['cgpa']}")

#Add new Key

student["email"] = "priya@gmail.com"
student["city"] = "Bengaluru"
print(f"Added email: {student['email']}")
print(f"Added city: {student['city']}")
print(f"Entire Dict: {student}")

#Remove a key

del student["passed"]
print(f"Entire Dict: {student}")

removed = student.pop("college")
print(f"Removed college : {removed}")

print(f"Final Student Dict: {student}")

```

```python
scores = {"Maths": 88, "Science": 92, "English": 79, "History": 85}

print(f"Subjects: {list(scores.keys())}")
print(f"Values: {list(scores.values())}")
print(f"All {list(scores.items())}")

#stats Calculation
print(f"Higest: {max(scores.values())}")
print(f"Higest: {min(scores.values())}")
print(f"Higest: {sum(scores.values())}")
print(f"Average: {sum(scores.values())/len(scores)}")


extra = {"Computer": 96, "PE": 90}
scores.update(extra)
print(f"After update{scores}")

print(type(scores))

```

```python
#Nested Dictionary

# Dictionary --> Dictonary
students_db = {
    "STU001": {
        "name"  : "Rahul Gupta",
        "marks" : {"Maths": 88, "Science": 92, "English": 79},
        "grade" : "A",
    },
    "STU002": {
        "name"  : "Sneha Patel",
        "marks" : {"Maths": 95, "Science": 97, "English": 91},
        "grade" : "A+",
    },
    "STU003": {
        "name"  : "Arjun Singh",
        "marks" : {"Maths": 62, "Science": 70, "English": 68},
        "grade" : "B",
    }
}


# print(f"{students_db["STU001"]["name"]}: {students_db["STU001"]["marks"]["Maths"]}")
# print(students_db["STU002"]["marks"]["Maths"])
# print(students_db["STU002"]["marks"].get("History","N/A"))


print("Student Report Card")

for student_id, info in students_db.items():
    name = info["name"]
    marks = info["marks"]
    grade = info["grade"]

    total = sum(marks.values())
    average = total/ len(marks)
    
    print(f" \n{student_id} - {name} [{grade}]")
    for subject, score in marks.items():
        bar = "❚" * (score // 10)
        print(f"{subject}: {score}  {bar}")
    print(f"\n Average :{average:.1f}")




```
