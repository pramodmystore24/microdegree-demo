```python
# =============================================
# 🔍 LIST INDEXING — GETTING ITEMS BY POSITION
# =============================================

order = ["Samosa", "Pav Bhaji", "Chole Bhature", "Lassi", "Jalebi"]
#         [0]         [1]             [2]           [3]      [4]
#         [-5]       [-4]             [-3]          [-2]     [-1]

print("First item :", order[0])      # Samosa
print("Last item  :", order[-1])     # Jalebi
print("3rd item   :", order[2])      # Chole Bhature

# Slicing — get a range
print("First 3    :", order[0:3])    # ['Samosa', 'Pav Bhaji', 'Chole Bhature']
print("Last 2     :", order[-2:])    # ['Lassi', 'Jalebi']
print("Every other:", order[::2])    # ['Samosa', 'Chole Bhature', 'Jalebi']
print("Reversed   :", order[::-1])   # ['Jalebi', 'Lassi', ...]

# Check if item exists
item_to_check = "Lassi"
if item_to_check in order:
    print(f"✅ {item_to_check} is in your order!")
else:
    print(f"❌ {item_to_check} not found in order.")

# Find position of an item
position = order.index("Pav Bhaji")
print(f"Pav Bhaji is at position: {position}")
```

```python
#Modifying a list
# add, Remove, Update
#Loop in list
#List Functions
#Tuples - 
#Reading a Tuple
#Random module
#time

cart = ["Burger", "Fries", "Coke"]
print(f"Initial cart {cart}")

#Adding
cart.append("Ice Cream") #Add items to the end
print(f"After Append:  {cart}")

cart.insert(1, "Garlic  Bread")
print(f"After Insert:  {cart}")

another_order = ["Salad", "Soup"] #Add Multiple items to another list
cart.extend(another_order)
print(f"After Extend:  {cart}")

#Remove 

cart.remove("Salad")  #Removes only the first match
print(f"After Remove:  {cart}")

removed_item = cart.pop() #remove the last item and return the last item 
print(f"Removed: {removed_item}")
print(f"After Pop:  {cart}")

removed_item = cart.pop(0)
print(f"Removed with index: {removed_item}")
print(f"After Pop with index:  {cart}")

#Update
cart[0] = "Veggie Burger"
print(f"After Update:  {cart}")


```

```python
# Features:
# 1. Start with an empty watchlist = []
# 2. Show a menu:
#    a) Add movie
#    b) Remove movie
#    c) Mark as watched (move from watchlist to watched_list)
#    d) Show all (watchlist + watched count)
#    e) Show stats (total, watched, remaining)
#    f) Sort watchlist alphabetically
#    g) Exit

# 3. Use separate lists: watchlist = [] and watched = []
# 4. When user types a movie name to remove, handle if it's not found
# 5. Show a recommendation: "You've watched X movies! Great job!"

# Sample output:
#   🎬 WATCHLIST MANAGER
#   ─────────────────────
#   Pending  (3): Inception, Interstellar, The Dark Knight
#   Watched  (2): Parasite, 3 Idiots
  
#   Stats: 5 total | 2 watched | 3 remaining
```

```python
watchlist = []
watched = []

while True:
    
    print("\n Watchlist Manager")
    print("-" * 30)
    print("1. Add movie")
    print("2. Remove movie")
    print("3. Mark as Watched")
    print("4. Show All")
    print("5. Show Stats")
    print("6. Sort Watchlist")
    print("7. Exit")


    choice = input("\n Enter your choice: ")

    #Add movie
    if choice == "1":
        movie = input("Enter Movie Name: ").lower()
        if movie not in watchlist and movie not in watched:
            watchlist.append(movie)
            print(f"{movie} Movie successfully added to the list")
        else:
            print(f"{movie} already present")

    #Remove Movie
    elif choice == "2":
        movie = input("Enter Movie Name: ").lower()

        if movie in watchlist:
            watchlist.remove(movie)
            print(f"{movie} is removed")
        else:
            print("Movie is not found")
    
    #Mark as watched
    elif choice == "3":
        movie = input("Enter Movie Name which you have already watched: ").lower()
        if movie in watchlist:
            watchlist.remove(movie)
            watched.append(movie)
            print(f"{movie} is updated in watched")
        else:
            print("Movie is not found in watchlist")

    # Show all

    elif choice == "4":
        print("Your Watch list")
        if watchlist:
            print(f"Pending ({len(watchlist)}): {', '.join(watchlist)}")
        else:
            print("Pending : None")

        if watched:
            print(f"Watched ({len(watched)}): {', '.join(watched)}")
        else:
            print("Watched : None")

    #Show Stats 
    elif choice == "5":

        total = len(watched) +len(watchlist)

        print("Stats")
        print(f"Total Movies: {total}")
        print(f"Pending : {len(watchlist)}")
        print(f"Watched : {len(watched)}")
        print(f"Congrats! you have finsihed watching so many movies")

    #Sort Watchlist

    elif choice == "6":
        watchlist.sort()
        print(f"Watchlist Sorted")
        print(", ".join(watchlist))

    elif choice == "7":
        print("Thanks for using watchlist manager")
        break
    
    else:
        print("Invalid Choice")
        break
        



```
